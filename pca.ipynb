Principal Component Analysis (PCA)


```python
# Loading libraries
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loading csv data (combined.csv) into dataframe
df = pd.read_csv('data/combined.csv')
attributes = df.columns.tolist()

# Removing unnecessary column that holds index
df = df.drop(labels=['Unnamed: 0'],axis=1)
```

```python
# Describing the dataframe
print(df.head(10))
print(df.shape)
print(attributes)
```

```python
# Finding correlation matrix
df.corr()

# Visualizing correlation matrix
cm = df.corr()
plt.figure(figsize=(15,15))
sns.heatmap(cm,vmax=1,square=True,annot=True,cmap='cubehelix')
plt.title('Correlation between different ride attributes')
```
